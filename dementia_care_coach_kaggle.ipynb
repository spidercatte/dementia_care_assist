## The Problem Nobody Talks About

Over **55 million people** worldwide live with dementia. Most are cared for at home — not in clinical facilities — by a single family member who stepped into the role without training and without a realistic picture of how consuming it would become.

The standard advice is to share the load: hire a Personal Support Worker, arrange respite care. The logic is sound. **The reality is devastating.**

Dementia profoundly changes how the brain processes trust and familiarity. Patients in moderate to advanced stages frequently experience heightened paranoia and severe anxiety around unfamiliar faces. When a new face enters the home — even a warm, trained, well-intentioned one — the patient's brain may register that person as an intruder.

**This creates a trap families cycle through for years:**

```
Primary caregiver burns out
        ↓
Tries to bring in support
        ↓
New caregiver's face triggers patient distress
        ↓
Primary caregiver feels guilty → refuses more help
        ↓
Burnout accelerates
        ↓
        (repeat)
```

There is no always-available, personalized coach standing beside a caregiver at 2am. And there is no way for every new person entering the care circle to arrive already knowing the patient.

**No existing tool addresses this transition problem directly.**

## What We Built

**DementiaCare Coach** is a multi-agent AI system designed to solve two problems simultaneously:

1. **Coach caregivers through difficult interactions** — in real time, grounded in clinical evidence
2. **Onboard every new person who enters the care circle** — so new caregivers arrive informed, not cold

The platform **learns the patient**. Over time, through profile setup and automatic enrichment from real coaching conversations, the system builds a detailed picture — not just clinically, but as a person: the music they respond to, the phrases that calm them, the topics that trigger distress, the approaches that have worked.

**When a new PSW starts, they do not arrive cold.**

### Why a Pipeline Instead of One Big Prompt?

| Single LLM | 7-Agent Pipeline |
|---|---|
| Safety flags buried in prose | Dedicated SafetyEvaluator — always runs, always surfaced |
| Generic advice | RAG-grounded clinical protocols per behavior type |
| One-size-fits-all | PatientContextProcessor maps patient profile against observed behavior |
| Hallucination hard to isolate | One agent, one responsibility — failures are diagnosable per step |
| No memory | Persistent interaction history + HITL profile learning |

## The 7-Agent Pipeline in Detail

Each step is a dedicated Gemini agent with a narrow system prompt and a **Pydantic-enforced JSON output schema** (`response_mime_type="application/json"`). This eliminates free-form hallucination and makes every step's output machine-readable for the next agent.

| Step | Agent | Key Responsibility |
|---|---|---|
| 0 | `ValidationService` | Reject noise before spending tokens — silence, unrelated media, greetings with no behavioral signal |
| 1 | `InteractionAnalyzer` | Extract agitation level (1–5), behavioral timeline, and a RAG keyword from text, audio, or video input |
| 2 | `PatientContextProcessor` | Map the patient's documented triggers and preferences against what was observed |
| RAG | Vector Store | Retrieve top-2 clinical protocols matching the Step 1 keyword from 20+ indexed guidelines |
| 3 | `CareGuidanceService` | Synthesize RAG output into recommendations specific to this behavior and patient |
| 4 | `SafetyEvaluator` | Dedicated safety pass — fall risk, delirium signs, medication omission; HIGH/EMERGENCY flags persisted via MCP |
| 5 | `CoachingSynthesizer` | Assemble the caregiver-facing response: strengths, changes, Try/Avoid dialog scripts |
| Post | `ProfileEnricherAgent` | Detect new triggers/preferences mentioned in coach chat; surface as HITL suggestions — no DB write without caregiver approval |

````python
# The orchestrator wires each agent's output to the next agent's input.
# This is a simplified view of the pipeline from backend/app/agents/orchestrator.py

PIPELINE_STEPS = [
    {
        "step": 0,
        "agent": "ValidationService",
        "purpose": "Intake gatekeeper — reject noise before spending tokens on analysis",
        "output_schema": "ValidationResponse(is_valid, reason, content_type)",
        "stops_pipeline_if": "not is_valid"
    },
    {
        "step": 1,
        "agent": "InteractionAnalyzer",
        "purpose": "Extract behavioral signals — agitation level, timeline, RAG query keyword",
        "output_schema": "InteractionAnalysisResponse(agitation_level, primary_behavior, behavioral_timeline, rag_query)",
        "multimodal": "Receives video/audio via Gemini File API when media is submitted"
    },
    {
        "step": 2,
        "agent": "PatientContextProcessor",
        "purpose": "Map patient profile (history, triggers, preferences) against observed behavior",
        "output_schema": "PatientContextResponse(relevant_triggers, violated_preferences, historical_context)",
        "data_source": "PostgreSQL patient profile via MCP get_patient_profile tool"
    },
    {
        "step": "RAG",
        "agent": "Vector Store",
        "purpose": "Retrieve top-2 clinical guidelines matching Step 1 rag_query",
        "production": "Vertex AI Search",
        "local": "ChromaDB + Gemini text-embedding-004",
        "indexed_documents": "20+ clinical protocol markdown files (medication refusal, sundowning, wandering, aggression, etc.)"
    },
    {
        "step": 3,
        "agent": "CareGuidanceService",
        "purpose": "Synthesize RAG results into actionable clinical recommendations for this specific scenario",
        "output_schema": "CareGuidanceResponse(recommended_approaches, approaches_to_avoid, clinical_rationale)"
    },
    {
        "step": 4,
        "agent": "SafetyEvaluator",
        "purpose": "Dedicated safety pass — fall risk, delirium, medication omission. Never buried in coaching output.",
        "output_schema": "SafetyEscalationResponse(risk_level, escalation_required, fall_risk, delirium_indicators)",
        "side_effect": "HIGH/EMERGENCY flags written to PostgreSQL via MCP log_safety_escalation tool"
    },
    {
        "step": 5,
        "agent": "CoachingSynthesizer",
        "purpose": "Assemble final caregiver response from all previous outputs",
        "output_schema": "FinalCoachingResponse(what_you_did_well, what_to_change, dialog_scripts, clinical_grounding)"
    },
    {
        "step": "Post (async)",
        "agent": "ProfileEnricherAgent",
        "purpose": "Human-in-the-Loop: detect new triggers/preferences in coach chat; surface for caregiver approval",
        "output_schema": "ProfileEnrichmentResponse(suggestions=[{field, value, confidence, source_evidence}])",
        "hitl_mechanism": "Suggestions shown in frontend — caregiver approves/rejects before any DB write"
    }
]

print("DementiaCare Coach — Agent Pipeline")
print("=" * 70)
for step in PIPELINE_STEPS:
    step_label = str(step['step'])
    print(f"\n[Step {step_label}] {step['agent']}")
    print(f"  Purpose: {step['purpose']}")
    if 'output_schema' in step:
        print(f"  Output:  {step['output_schema']}")
    if 'hitl_mechanism' in step:
        print(f"  HITL:    {step['hitl_mechanism']}")
    if 'side_effect' in step:
        print(f"  Side effect: {step['side_effect']}")
````

## The MCP Server & ADK Agent

### MCP Server
A standalone **FastMCP** service over Server-Sent Events (SSE) at port 8002. Four tools, consumed by both the FastAPI pipeline and the external ADK conversational agent:

```python
# mcp_server/app/mcp_server.py (simplified)
@mcp.tool()
async def get_patient_profile(patient_name: str) -> dict:
    """Load full patient profile from PostgreSQL."""
    ...

@mcp.tool()
async def log_safety_escalation(patient_name: str, risk_level: str, details: str) -> dict:
    """Persist HIGH/EMERGENCY safety flags for clinician review."""
    ...

@mcp.tool()
async def query_care_guidelines(query: str, n_results: int = 2) -> list:
    """RAG retrieval from Vertex AI Search (prod) / ChromaDB (local)."""
    ...

@mcp.tool()
async def search_patients(query: str) -> list:
    """Search patient records — used by ADK agent to find patient context."""
    ...
```

All data access and side effects route through these MCP tools. This keeps the ADK agent **fully stateless and independently deployable** — it has no direct database connection.

### ADK Conversational Agent
A `google.adk.agents.Agent` in `adk-agent-scaffold/` connected to the backend **exclusively via MCP**. Each conversation follows a strict protocol:
1. Load patient profile via `get_patient_profile`
2. Retrieve relevant clinical context via `query_care_guidelines`
3. Log any safety concerns via `log_safety_escalation`
4. Synthesize a compassionate, specific, actionable response

Chat history is persisted per patient and reloaded across sessions — a caregiver can return two days later and the coach already knows what they discussed.

**Deployed to:** Vertex AI Reasoning Engine via `agents-cli deploy`

## Human-in-the-Loop: Profile Enrichment

The **ProfileEnricherAgent** is DementiaCare Coach's continuous learning mechanism. After every coach chat exchange, it runs asynchronously to detect new information mentioned by the caregiver:

- *"She's been responding really well to Frank Sinatra in the mornings"* → suggests adding `music_preferences: ["Frank Sinatra"]`
- *"Dad always gets worse after large family gatherings"* → suggests adding `behavioral_triggers: ["large social gatherings"]`
- *"He refused again when I mentioned the hospital"* → suggests adding `distress_topics: ["hospital"]` with confidence 0.85

Each suggestion includes:
- **Field** — which patient profile attribute to update
- **Value** — the proposed addition
- **Confidence** — 0.0–1.0 based on how clearly it was stated
- **Source evidence** — the exact quote from the conversation that triggered the suggestion

The caregiver sees these as reviewable cards in the frontend — **approve** to write to the profile, **dismiss** to discard. No profile change occurs without explicit consent. The patient's care knowledge grows more accurate with every real interaction.

## Privacy, Ethics & Patient Consent

Recording a vulnerable patient in a moment of distress is a high-stakes privacy problem. DementiaCare Coach implements a concrete consent governance framework:

**1. Surrogate Consent Gating**
Video and audio analysis requires documented surrogate consent from the patient's legal decision-maker (Power of Attorney, Legal Guardian), scoped per media type, before any file is uploaded to an external API. Consent is stored in the `consent_records` table and checked at the orchestrator intake layer (Step 0).

**2. Auditable Trail**
Every consent verification check is logged to `consent_audit_logs` — who authorized, required scope, allowed scope, result, timestamp. Clinical-grade audit record.

**3. No Persistent Media Storage**
Raw video, audio, and images are **never permanently stored**:
- Files are buffered in `tempfile.NamedTemporaryFile` for the API call duration and deleted in a `finally` block
- Gemini File API uploads are deleted via `client.files.delete(...)` immediately after analysis
- A background task purges orphaned files every 15 minutes
- Google's 48-hour absolute expiry acts as a final safety net

**4. Text-Only Alternative**
Caregivers can always describe interactions in writing. No recording, no upload required.

Full threat model: [`threat_model.md`](https://github.com/spidercatte/dementia_care_assist/blob/main/threat_model.md) — covers STRIDE threat categories across all system components.

The trap — where trying to get help makes things worse — has a way out.
